```python
import pandas as pd
import numpy as np
from datetime import datetime

data = pd.read_excel('Blue_Bucket_data_2015-Aug2021.xlsx', sheet_name=None)
```

```python
data.pop('Source Data', None)
data.pop('Pivot Table', None)
data.keys()
```

```python
# type(data['2021 blue bucket data'].iloc[8, 5]), data['2021 blue bucket data'].iloc[8, 5]

data['2021 blue bucket data'] = data['2021 blue bucket data'].drop([0, 1])
```

```python
for x in data['2021 blue bucket data'].columns:
#     print(type(x), type(data['2021 blue bucket data'].iloc[0].loc[x]))
    data['2021 blue bucket data'].rename(columns={str(x): str(data['2021 blue bucket data'].iloc[0].loc[x])}, inplace=True)
```

```python
data['2021 blue bucket data'] = data['2021 blue bucket data'].drop([2])
```

```python
locs = []
for index, row in data['2021 blue bucket data'].iterrows():
    a = (type(row['Monthly Cleanup Site']) != str)
    b = (type(row['Habitat Restoration Site']) != str)
    c = (type(row['Event Cleanup Site']) != str)
    d = (type(row['Individual Cleanup Location']) != str)
    if not a:
        locs.append(row['Monthly Cleanup Site'])
    if not b:
        locs.append(row['Habitat Restoration Site'])
    if not c:
        locs.append(row['Event Cleanup Site'])
    if not d:
        locs.append(row['Individual Cleanup Location'])
    if a and b and c and d:
        data['2021 blue bucket data'] = data['2021 blue bucket data'].drop([index])
```

```python
import string
locs2 = []
for s in locs:
    s = s.lower()
    s = ' '.join(word.strip(string.punctuation) for word in s.split())
    locs2.append(s)

locs2
```

```python
locs2[0].split(" ", 1)
```

```python
import difflib as dl

for i in locs2:
    i.split(" ")
substring_counts={}

for i in range(0, len(locs)):
    for j in range(i+1,len(locs)):
        string1 = locs[i]
        string2 = locs[j]
        match = dl.SequenceMatcher(None, string1, string2).find_longest_match(0, len(string1), 0, len(string2))
        matching_substring=string1[match.a:match.a+match.size]
        if(matching_substring not in substring_counts):
            substring_counts[matching_substring]=1
        else:
            substring_counts[matching_substring]+=1

print(substring_counts)

import operator
max_occurring_substring=max(substring_counts.iteritems(), key=operator.itemgetter(1))[0]
print(max_occurring_substring)
```
