```python
import pandas as pd
import numpy as np


data = pd.read_csv("Tides_and_Blue_Bucket_Data.csv")
```

```python
df = data
df["Total Recyclables"] = df['Take Out/Away Containers (Plastic)'] + df['Bottle Caps (Plastic)'] + df['Lids (Plastic)'] + df['Beverage Bottles (Plastic)'] + df['Beverage Bottles (Glass)'] + df['Beverage Cans']+ df['Grocery Bags (Plastic)'] + df['Other Plastic Bags'] + df['Paper Bags'] + df['Cups, Plates (Paper)']+ df['Cups, Plates (Plastic)'] + df['6-Pack Holders'] + df['Other Plastic Bottles (oil, bleach, etc.)'] + df['Strapping Bands'] + df['Tobacco Packaging/Wrap'] + df['Fishing Net & Pieces'] + df['Rope (1 yard/meter = 1 piece)']+ df['Tires'] + df['Plastic Pieces']
df.drop(['Unnamed: 0'],axis =1)
df["Total Recyclables"]
```

```python
data['Cleanup Timestamp'] = pd.to_datetime(data['Cleanup Date'], format="%Y-%m-%d %H:%M:%S")
```

```python
df = data.groupby('Cleanup Timestamp').agg('sum')
df.reset_index(level=0, inplace=True)
```

```python
df['Cumulative Recyclables'] = df['Total Recyclables'].cumsum()
```

```python
import plotly.express as px

fig = px.line(df, x="Cleanup Timestamp", y="Cumulative Recyclables")
fig.show()
```

```python
# df2 = df.groupby('Month, Year').agg('sum')
per_month = df.set_index('Cleanup Timestamp').groupby(pd.Grouper(freq='M'))['Total Recyclables'].sum()

x = per_month.index
y = per_month.values

df2 = pd.DataFrame({'Month': x, 'Total Recyclables': y})

fig = px.line(df2, x='Month', y='Total Recyclables')
fig.show()
```
