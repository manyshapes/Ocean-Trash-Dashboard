```python
locs = []
for index, row in data['2021 blue bucket data'].iterrows():
    a = (type(row['Monthly Cleanup Site']) != str)
    b = (type(row['Habitat Restoration Site']) != str)
    c = (type(row['Event Cleanup Site']) != str)
    d = (type(row['Individual Cleanup Location']) != str)
    if not a:
        locs.append(row['Monthly Cleanup Site'])
    if not b:
        locs.append(row['Habitat Restoration Site'])
    if not c:
        locs.append(row['Event Cleanup Site'])
    if not d:
        locs.append(row['Individual Cleanup Location'])
    if a and b and c and d:
        data['2021 blue bucket data'] = data['2021 blue bucket data'].drop([index])
```
