```python
#Reading in Datasets
Tides_Data= pd.read_csv("TIDES_2015-2021.csv")
Blue_Bucket_Data = pd.read_excel("Blue_Bucket_data_2015-Aug2021.xlsx")
```

```python
#After reading the columns of the Blue Bucket dataset, we realized that Blue Buckets has the same column names as
#the Tides dataset except for 4 additional column names which have been listed in this cell.
"Year" in Tides_Data, "Month" in Tides_Data, "Unnamed: 67" in Tides_Data, "Unnamed: 68" in Tides_Data
```

```python
#Two of those columns are not needed so they can be removed from the dataset
Blue_Bucket_Data_cleaned = Blue_Bucket_Data.drop(columns=["Unnamed: 67", "Unnamed: 68"])
```

```python
#The other two columns that Blue Bucket has, "Month" and "Year", actually provide some useful information, but
#the Tides dataset does not have these columns, so I extracted the month and year from the "Cleanup Date" column
Tides_Data["Month"] = pd.DatetimeIndex(Tides_Data['Cleanup Date']).month
Tides_Data["Year"] = pd.DatetimeIndex(Tides_Data['Cleanup Date']).year
```

```python
#Showing the shapes of the modified Tides and Blue Bucket dataframes with the same number of columns now
print("Tide's shape: " + str(Tides_Data.shape)) 
print("Blue Bucket's shape: " + str(Blue_Bucket_Data_cleaned.shape))
print("Number of columns in Tides: " + str(Tides_Data.shape[1]))
print("Number of columns in Blue Bucket: " + str(Blue_Bucket_Data_cleaned.shape[1]))
```

```python
#Merging the Tides and Blue Bucket dataframes
Tides_and_Blue_Bucket = Tides_Data.append(Blue_Bucket_Data_cleaned, sort=False)
```

```python
#Introduce a new column called "Clean Timestamp" to the dataset that cleans up the format of the already existing "Cleanup Date" column
Tides_and_Blue_Bucket['Cleanup Timestamp'] = pd.to_datetime(Tides_and_Blue_Bucket['Cleanup Date'], format="%Y-%m-%d %H:%M:%S")
Tides_and_Blue_Bucket
```

```python
#Create a new column called "Single Use Plastics" by binning data off of already existing columns that classify as single use plastics 
Tides_and_Blue_Bucket['Single Use Plastics'] = Tides_and_Blue_Bucket['Cigarette Butts'] + Tides_and_Blue_Bucket['Food Wrappers (candy, chips, etc.)'] + Tides_and_Blue_Bucket['Take Out/Away Containers (Plastic)'] + Tides_and_Blue_Bucket['Bottle Caps (Plastic)'] + Tides_and_Blue_Bucket['Lids (Plastic)'] + Tides_and_Blue_Bucket['Straws, Stirrers'] + Tides_and_Blue_Bucket['Beverage Bottles (Plastic)'] + Tides_and_Blue_Bucket['Grocery Bags (Plastic)'] + Tides_and_Blue_Bucket['Other Plastic Bags'] + Tides_and_Blue_Bucket['Cups, Plates (Plastic)'] + Tides_and_Blue_Bucket['Fishing Net & Pieces'] + Tides_and_Blue_Bucket['Fishing Line (1 yard/meter = 1 piece)'] + Tides_and_Blue_Bucket['6-Pack Holders'] + Tides_and_Blue_Bucket['Balloons'] + Tides_and_Blue_Bucket['Beverages Sachets'] + Tides_and_Blue_Bucket['Cigarette Lighters'] + Tides_and_Blue_Bucket['Tobacco Packaging/Wrap'] + Tides_and_Blue_Bucket['Diapers'] + Tides_and_Blue_Bucket['Tampons/Tampon Applicators'] + Tides_and_Blue_Bucket['Other Plastic/Foam Packaging'] + Tides_and_Blue_Bucket['Other Plastic Bottles (oil, bleach, etc.)'] + Tides_and_Blue_Bucket['Gloves & Masks (PPE)']
Tides_and_Blue_Bucket['Single Use Plastics']
```

```python
#Create a new column called "Total Recyclables" by binning data off of already existing columns that classify as recyclable items
Tides_and_Blue_Bucket["Total Recyclables"] = Tides_and_Blue_Bucket['Take Out/Away Containers (Plastic)'] + Tides_and_Blue_Bucket['Bottle Caps (Plastic)'] + Tides_and_Blue_Bucket['Lids (Plastic)'] + Tides_and_Blue_Bucket['Beverage Bottles (Plastic)'] + Tides_and_Blue_Bucket['Beverage Bottles (Glass)'] + Tides_and_Blue_Bucket['Beverage Cans']+ Tides_and_Blue_Bucket['Grocery Bags (Plastic)'] + Tides_and_Blue_Bucket['Other Plastic Bags'] + Tides_and_Blue_Bucket['Paper Bags'] + Tides_and_Blue_Bucket['Cups, Plates (Paper)']+ Tides_and_Blue_Bucket['Cups, Plates (Plastic)'] + Tides_and_Blue_Bucket['6-Pack Holders'] + Tides_and_Blue_Bucket['Other Plastic Bottles (oil, bleach, etc.)'] + Tides_and_Blue_Bucket['Strapping Bands'] + Tides_and_Blue_Bucket['Tobacco Packaging/Wrap'] + Tides_and_Blue_Bucket['Fishing Net & Pieces'] + Tides_and_Blue_Bucket['Rope (1 yard/meter = 1 piece)']+ Tides_and_Blue_Bucket['Tires'] + Tides_and_Blue_Bucket['Plastic Pieces']
Tides_and_Blue_Bucket['Total Recyclables']
```

```python
# Create a new dataframe that groups rows of the Tides_and_Blue_Bucket dataframe together by date 
# so that numerical data of all trash items that were collected on certain date is summed up together.
# However, some columns such as Month and Year become useless as a result so they must be dropped.
Tides_and_Blue_Bucket_group_by_date = Tides_and_Blue_Bucket.groupby('Cleanup Timestamp').agg('sum')
Tides_and_Blue_Bucket_group_by_date.reset_index(level=0, inplace=True)
Tides_and_Blue_Bucket_group_by_date.drop(columns=["Year", "Month"])
```

```python
#Create a column that tracks the amount of single use plastics were cumulatively collected overtime in chronological order.
Tides_and_Blue_Bucket_group_by_date['Cumulative Single Use Plastics'] = Tides_and_Blue_Bucket_group_by_date['Single Use Plastics'].cumsum()
Tides_and_Blue_Bucket_group_by_date['Cumulative Single Use Plastics']
```

```python
#Group the merged Tides and Blue Bucket dataset by Zone and summed the Pounds column for each group 
#(I couldn't find a column for site location cleanup so this was the closest thing to location)
#Then get the top 10 zones with the largest weights of trash

Tides_and_Blue_Bucket_grouped = Tides_and_Blue_Bucket.groupby('Zone')['Pounds'].sum().nlargest(10)
```

```python
#Contents of new series after grouping
Tides_and_Blue_Bucket_grouped
```

```python
#Convert the series into a dataframe so pounds can be sorted in descending order
Tides_and_Blue_Bucket_grouped_df = pd.DataFrame()
Tides_and_Blue_Bucket_grouped_df['Zone'] = Tides_and_Blue_Bucket_grouped.index
Tides_and_Blue_Bucket_grouped_df['Total Pounds'] = Tides_and_Blue_Bucket_grouped.values
```

```python
#Contents of new dataframe
Tides_and_Blue_Bucket_grouped_df
```

```python
#Sort the entries of dataframe by total pounds in descending order
Tides_and_Blue_Bucket_sorted = Tides_and_Blue_Bucket_grouped_df.sort_values(by = 'Total Pounds', ascending = False)
Tides_and_Blue_Bucket_sorted
```

```python

fig = go.Figure(data=[go.Table(
    header=dict(values=['Zone', 'Total Pounds'],
                line_color='darkslategray',
                fill_color='lightskyblue',
                align='left'),
    cells=dict(values=[Tides_and_Blue_Bucket_sorted['Zone'], # 1st column
                       Tides_and_Blue_Bucket_sorted['Total Pounds']], # 2nd column
               line_color='darkslategray',
               fill_color='lightcyan',
               align='left'))
])

fig.update_layout(width=600, height=300)
fig.show()
```
